```python
import gensim

from gensim.corpora.dictionary import Dictionary
from gensim.utils import simple_tokenize
from gensim.models import LdaMulticore

from the_pile.datasets import *
```

```python
dataset = ExPorterDataset()
```

```python
dataset._download()
```

```python
dataset.size()
```

```python
dictionary = Dictionary()
```

```python
docs_left = dataset.num_docs()
dictionary_docs = dataset.documents()
dictionary.add_documents([simple_tokenize(next(dictionary_docs)[0]) for _ in range(1000)], prune_at=10000)
```

```python
BATCH_SIZE = 1000
BATCH_SIZE = min(BATCH_SIZE, docs_left)
```

```python
corpus_docs = dataset.documents()
print("DOCUMENTS LEFT: ", docs_left)
base_corpus = [dictionary.doc2bow(simple_tokenize(next(corpus_docs)[0])) for _ in range(0, BATCH_SIZE)]
docs_left -= BATCH_SIZE
lda = LdaMulticore(base_corpus, id2word=dictionary, num_topics=10)
print("DOCUMENTS LEFT: ", docs_left)
```

```python
while docs_left > BATCH_SIZE:
    other_corpus = [dictionary.doc2bow(simple_tokenize(next(corpus_docs)[0])) for _ in range(0, BATCH_SIZE)]
    docs_left -= BATCH_SIZE
    lda.update(other_corpus)
    print("DOCUMENTS LEFT: ", docs_left)
    
other_corpus = [dictionary.doc2bow(simple_tokenize(next(corpus_docs)[0])) for _ in range(0, docs_left)]
docs_left = 0
if len(other_corpus) > 0:
    lda.update(other_corpus)
print("DOCUMENTS LEFT: ", docs_left)
```
